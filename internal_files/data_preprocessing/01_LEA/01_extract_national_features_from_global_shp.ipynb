```python
countries = countrydata.Political.get_union('ECOWAS')
```

```python
# load all country shapes
gid_shp="/storage/internal/data/gears/lecture_heidi_niger/data/00_general_data/00_geospatial_data/GADM_country_shapes/gadm36_0.shp"
countries_str="','".join(countries)
country_df=gk.vector.extractFeatures(
    source=gid_shp,
    where=f"GID_0 in ('{countries_str}')"
)
country_df
```

```python
for country in countries:
    if not os.path.exists(os.path.join(output_folder, country)):
        os.mkdir(os.path.join(output_folder, country))
    df = gk.vector.extractFeatures(
        source=input_filepath,
        geom=country_df[country_df.GID_0==country].geom.iloc[0]
    ).reset_index(drop=True)
    if len(df)==0:
        with open( os.path.join(output_folder, country, output_filename.replace('COUNTRY', country))[:-3]+'txt', "w") as file:
            file.write(f"No features found for country {country} in file: {os.path.basename(input_filepath)}")
        print("txt spacer saved for", country)
    else:
        gk.vector.createVector(
            df,
            os.path.join(output_folder, country, output_filename.replace('COUNTRY', country)),
        )
```
