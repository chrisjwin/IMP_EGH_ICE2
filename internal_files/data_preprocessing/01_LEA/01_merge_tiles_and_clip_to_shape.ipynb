```python
import geokit as gk
import countrydata
import datetime
import os
import pathlib
```

```python
output_basefolder=os.path.join(os.path.join(pathlib.Path().cwd().parent.parent.parent),"student_files/data/exclusion_datasets/WSF2019")
output_basefolder
```

```python
# load all country shapes
gid_shp="/storage/internal/data/c-winkler/04_Data/01_Geodata/Political_Geography/01_Country_Borders/ALL_global/gadm36_0.shp"

countries=countrydata.Political.get_union("ECOWAS")
countries_str="','".join(countries)

country_df=gk.vector.extractFeatures(
    source=gid_shp,
    where=f"GID_0 in ('{countries_str}')"
)
country_df
```

```python
tile_basepath="/storage/internal/data/c-winkler/04_Data/01_Geodata/Infrastructure/01_Settlements/WSF2019/download.geoservice.dlr.de/WSF2019/files/WSF2019_v1_*.tif"
```

```python
for country in countries:
    print(datetime.datetime.now(), "Now processing", country)
    os.mkdir(os.path.join(output_basefolder, country))
    geom = country_df[country_df.GID_0==country].geom.iloc[0]
    sources_filterobject = gk.Extent.load(geom).filterSources(tile_basepath)
    sources_list = [s for s in sources_filterobject]
    raster_comb = gk._algorithms.combineSimilarRasters.combineSimilarRasters(
        datasets=sources_list, 
        output=os.path.join(output_basefolder, country, f"{country}_extract_WSF2019.tif"),
        )
print(datetime.datetime.now(), "ALL COMPLETE")
```
