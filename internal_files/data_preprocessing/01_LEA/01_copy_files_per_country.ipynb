```python
import geokit as gk
import countrydata
import shutil
import os
import glob
```

```python
input_basepath='/storage/internal/data/c-winkler/04_Data/02_Geodata_processed/Infrastructure/osm_data_power_lines/data/COUNTRY/powerlines_COUNTRY.shp'
output_folder='/storage/internal/home/c-winkler/Research/02_Projects/02_WASCAL_graduate_school/data/01_LEA_data/00_exclusion_datasets/OSM'
output_filename=f"COUNTRY_osm_powerlines.EXTENSION"
```

```python
# load all country shapes
gid_shp="/storage/internal/data/gears/lecture_heidi_niger/data/00_general_data/00_geospatial_data/GADM_country_shapes/gadm36_0.shp"

countries=countrydata.Political.get_union("ECOWAS")
countries_str="','".join(countries)

country_df=gk.vector.extractFeatures(
    source=gid_shp,
    where=f"GID_0 in ('{countries_str}')"
)
country_df
```

```python
for country in countries:
    print("Now", country)
    if not os.path.exists(os.path.join(output_folder, country)):
        os.mkdir(os.path.join(output_folder, country))
    for extension in [fp.split('.')[-1] for fp in glob.glob(input_basepath.replace("COUNTRY",country)[:-4]+'.*')]:
        shutil.copy(
            input_basepath.replace('COUNTRY', country)[:-3]+extension,
            os.path.join(output_folder, country, output_filename.replace("COUNTRY",country).replace("EXTENSION",extension))
        )
```
